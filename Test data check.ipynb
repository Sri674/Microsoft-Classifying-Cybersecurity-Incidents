```python
import pandas as pd

test_df=pd.read_csv("test_dataset.csv") 
test_df.head()
```

```python
test_df.drop(columns=["ActionGrouped","ActionGranular","EmailClusterId","ThreatFamily","ResourceType","Roles","AntispamDirection","SuspicionLevel","LastVerdict"],inplace=True)
```

```python
test_df.drop(columns=["Timestamp"],inplace=True)
```

```python
test_df.drop(columns=["MitreTechniques"],inplace=True)
```

```python
test_df.drop(columns=["Id","OrgId","IncidentId","AlertId","DetectorId","NetworkMessageId"],inplace=True)
```

```python
test_df.drop(columns=["OAuthApplicationId","RegistryValueName", "RegistryValueData","OSFamily","ResourceIdName","Url","OSVersion",  "RegistryKey", "ApplicationId","EvidenceRole","ApplicationName"],inplace=True)
```

```python
test_df.head()
```

```python
test_df.drop(columns=["Usage"],inplace=True)
```

```python
import pandas as pd 

cols=[col for col in test_df.columns if col!="IncidentGrade"]
cols.append("IncidentGrade") 

test_df=test_df[cols] 
test_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder 

le=LabelEncoder() 

test_df["IncidentGrade"]=le.fit_transform(test_df["IncidentGrade"])
test_df.head()
```

```python
import pandas as pd


columns_to_encode = [
    'AlertTitle', 'Category', 'EntityType', 'DeviceId', 'Sha256', 'IpAddress', 
    'AccountSid', 'AccountUpn', 'AccountObjectId', 'AccountName', 'DeviceName', 
    'FileName', 'FolderPath', 'CountryCode', 'State', 'City'
]


for column in columns_to_encode:
    column_mean = test_df.groupby(column)['IncidentGrade'].mean()
    
    
    test_df[column] = test_df[column].map(column_mean).round(3)


test_df.head()
```

```python
test_df.to_csv("test_df.csv",index=False)
```

```python
import pandas as pd
final_test_df= pd.read_csv("test_df.csv") 
final_test_df.head()
```

```python
                             #MODEL LOAD FROM TRAIN DATASET

import joblib

Best_train_model=joblib.load("best_model_rf.pkl") 
print(Best_train_model)
```

```python
X=final_test_df.drop(columns=["IncidentGrade"]) 
y=final_test_df["IncidentGrade"] 


y_pred_t=Best_train_model.predict(X) 
y_pred_t
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(y, y_pred_t)
print("Accuracy score:", accuracy)


print("Classification report:")
print(classification_report(y, y_pred_t))


print("Confusion matrix:")
print(confusion_matrix(y, y_pred_t))
```

```python
Random_forest is best model chosen from train dataset, when it is tested on unseen test data it gives a accuracy of 0.86 
Precision,recall,f1-score too comes with good numbers 
in confusion matrix: class0,class1,class2 correctly identifies their respective classes in large number.
```
