```python
import pandas as pd

test_df=pd.read_csv("test_dataset.csv") 
test_df.head()
```

```python
test_df.drop(columns=["ActionGrouped","ActionGranular","EmailClusterId","ThreatFamily","ResourceType","Roles","AntispamDirection"],inplace=True)
```

```python
test_df.drop(columns=["SuspicionLevel","LastVerdict"], inplace=True)
```

```python
test_df.drop(columns=["OrgId","EvidenceRole","Sha256","AccountUpn","AccountObjectId","RegistryValueName","RegistryValueData"],inplace=True)
```

```python
test_df.drop(columns=["OAuthApplicationId","ResourceIdName","State","City"],inplace=True)
```

```python
test_df.drop(columns=["AlertTitle","OSVersion"],inplace=True)
```

```python
test_df.drop(columns=["Id","IncidentId","AlertId","DetectorId","NetworkMessageId"],inplace=True)
```

```python
test_df.drop(columns=["ApplicationName","ApplicationId","RegistryKey","OSFamily"],inplace=True)
```

```python
test_df.drop(columns=["MitreTechniques"],inplace=True)
```

```python
test_df.head()
```

```python
test_df["Timestamp"]=pd.to_datetime(test_df["Timestamp"])
```

```python
test_df['hour'] = test_df['Timestamp'].dt.hour
test_df['day'] = test_df['Timestamp'].dt.day
test_df['month'] = test_df['Timestamp'].dt.month
test_df['weekday'] = test_df['Timestamp'].dt.weekday
```

```python
test_df.drop(columns=['Usage'],inplace=True)
```

```python
import pandas as pd 

cols=[col for col in test_df.columns if col!="IncidentGrade"]
cols.append("IncidentGrade") 

test_df=test_df[cols] 
test_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder 

le=LabelEncoder() 

test_df["IncidentGrade"]=le.fit_transform(test_df["IncidentGrade"])
test_df.head()
```

```python
                                           #TARGET ENCODING FOR CATEGORICAL COLUMNS

import pandas as pd


columns_to_encode = ['Category','EntityType','DeviceId', 'IpAddress', 'Url', 'AccountSid', 'AccountName', 'DeviceName', 'FileName', 'FolderPath', 'CountryCode']


for column in columns_to_encode:
    column_mean = test_df.groupby(column)['IncidentGrade'].mean()
    
    
    test_df[column] = test_df[column].map(column_mean).round(3)


test_df.head()
```

```python
import pandas as pd 


columns_to_encode= ['hour', 'day', 'month', 'weekday'] 

for column in columns_to_encode:
     column_mean = test_df.groupby(column)['IncidentGrade'].mean()
     
     
     test_df[column]=test_df[column].map(column_mean).round(3) 


test_df.head()
```

```python
test_df.to_csv("test_df.csv",index=False)
```

```python
import pandas as pd 

final_test_df=pd.read_csv("test_df.csv") 
final_test_df.head()
```

```python
                             #MODEL LOAD FROM TRAIN DATASET

import joblib

Best_train_model=joblib.load("best_model_rf.pkl") 
print(Best_train_model)
```

```python
X=final_test_df.drop(columns=["IncidentGrade"]) 
y=final_test_df["IncidentGrade"] 


y_pred_t=Best_train_model.predict(X) 
y_pred_t
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(y, y_pred_t)
print("Accuracy score:", accuracy)


print("Classification report:")
print(classification_report(y, y_pred_t))


print("Confusion matrix:")
print(confusion_matrix(y, y_pred_t))
```
